```python
#input pagina web https://thispointer.com/pandas-drop-dataframe-columns-based-on-nan-percentage/
# Either values equal or more than 20% with NaN values in columns
perc =  50

min_count =  int(((100-perc)/100)*df.shape[0] + 1)

df = df.dropna( axis= 1,
                how='any',
                thresh= min_count,
                subset= None,
                inplace= False)
```
