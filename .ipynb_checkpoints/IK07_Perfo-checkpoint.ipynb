```python
import pandas as pd 
import os

#settings
pd.options.display.max_columns = 500
```

```python
path = 'D:/26 BI ENGINEER/Pandas - Python/datas'
filename = 'IK07_MEASPOINT_50313329.XLSX'

fullpath = os.path.join(path, filename)
```

```python
df = pd.read_excel(fullpath)

df.head()
```

```python
print(df.columns)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
#Percentage of NaN values
(df.isna().sum()/len(df))*100
```

```python
df.shape
```

```python
#input pagina web https://thispointer.com/pandas-drop-dataframe-columns-based-on-nan-percentage/
# Either values equal or more than 20% with NaN values in columns
perc =  50

min_count =  int(((100-perc)/100)*df.shape[0] + 1)

df = df.dropna( axis= 1,
                how='any',
                thresh= min_count,
                subset= None,
                inplace= False)
```
